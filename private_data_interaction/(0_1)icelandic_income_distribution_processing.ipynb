```python
import pandas as pd
```

```python
income_df = pd.read_csv('icelandic_income_distribution.csv')

# Define a dictionary for min and max age limits
income_age_limits = {
    "16-19": (16, 19),
    "20-24": (20, 24),
    "25-29": (25, 29),
    "30-34": (30, 34),
    "35-39": (35, 39),
    "40-44": (40, 44),
    "45-49": (45, 49),
    "50-54": (50, 54),
    "55-59": (55, 59),
    "60-64": (60, 64),
    "65-69": (65, 69),
    "70-74": (70, 74),
    "75-79": (75, 79),
    "80 and over": (80, 100),
}

# Map the dictionary to create two new columns: Min Age and Max Age
income_df['Min Age'] = income_df['Age groups'].map(
    lambda x: income_age_limits[x][0])
income_df['Max Age'] = income_df['Age groups'].map(
    lambda x: income_age_limits[x][1])

# Save the processed DataFrame
income_df_postprocessed = income_df

display(income_df_postprocessed)

# Optionally, save to a new CSV file
income_df_postprocessed.to_csv(
    'icelandic_income_distribution_postprocessed.csv', index=False)
```
